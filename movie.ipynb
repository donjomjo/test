```python
#시간이 지나면 안되는 코드입니다.
#
import pandas as pd
import requests
from bs4 import BeautifulSoup

url = "https://movie.naver.com/movie/running/current.nhn"
req = requests.get(url, verify=False)
html = req.text
soup = BeautifulSoup(html, 'html.parser')
a=[]
for tag in soup.select('dl[class=info_txt1]'):
    sstr = ''
    for i in range(len(tag.text)):
        if i <4:
            continue
        if tag.text[i] == '	'or tag.text[i] == '\r':
            continue
        sstr += tag.text[i]
    b=[]
    b = [i for i in list(sstr) if i !='\n']
    b=''.join(b).replace('출연', '|').replace('개봉감독', '|').split('|')
    a.append(b)
a[11].insert(4,'없음')
a[23].insert(4,'없음')
a[54].insert(4,'없음')
a[60].insert(4,'없음')
a[93].insert(4,'없음')
a[114].insert(0,'없음')
a[89].insert(0,'없음')
#a[119].insert(4,'없음')
i=0
g=list(range(232))

for tag in soup.select('span[class=num]'):   
    g[i] = tag.text
    i = i+1
    
g.insert(53,'0.04')
g.insert(97,'0.01')
g.insert(101,'0.01')
g.insert(113,'0.01')
g.insert(127,'0.01')
g.insert(235,'0.01')
g.insert(237,'0.01')
g.insert(239,'0.01')
#g.insert(143,'0.01')
#g.insert(147,'0.01')
#g.insert(151,'0.01')
#g.insert(231,'0.01')
#g.insert(233,'0.01')
#g.insert(235,'0.01')
#g.insert(237,'0.01')
#g.insert(239,'0.01')

j = 0
for e in g:
    a[int(j/2)].append(e)
    j = j + 1
    
j = 0

for tag in soup.select('dt[class=tit]'):
    c = []
    c = tag.text.strip().split('\n')
    if(len(c)==1):
        a[j].append('전체 관람가')
        a[j].append(c[0])
        j=j+1
        continue
    a[j].append(c[0])
    a[j].append(c[1])
    j=j+1
df = pd.DataFrame([['n1','n2','n3','n4','n5','n6','n7','n8','n8']], columns=['제목','연령','장르','영화길이','개봉일자','감독','출연','평점','예매율'])
i=0        
for d in a:
    movie = pd.DataFrame({
        '제목':[d[8]],
        '연령':[d[7]],
        '장르':[d[0]],
        '영화길이':[d[1]],
        '개봉일자':[d[2]],
        '감독':[d[3]],
        '출연':[d[4]],
        '평점':[d[5]],
        '예매율':[d[6]],
    })
    #print(len(d))
    #break
    df = df.append(movie, ignore_index=True)   
#print(a[117])
```
